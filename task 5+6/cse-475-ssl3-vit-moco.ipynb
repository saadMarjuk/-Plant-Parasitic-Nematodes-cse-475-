```python
# ====================
# 1. IMPORTS
# ====================
import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.optim.lr_scheduler import CosineAnnealingLR
from torch.utils.data import DataLoader, Dataset, random_split
from torchvision import transforms, models
from PIL import Image
import matplotlib.pyplot as plt
import time
import warnings
warnings.filterwarnings('ignore')
from tqdm import tqdm
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import copy
import random

# Set random seeds
def set_seed(seed=42):
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False

set_seed(42)

# Device
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# ====================
# 2. DATASET
# ====================
class NematodeDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.classes = sorted(os.listdir(root_dir))
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}
        self.images = []
        self.labels = []
        
        for class_name in self.classes:
            class_dir = os.path.join(root_dir, class_name)
            if os.path.isdir(class_dir):
                for img_name in os.listdir(class_dir):
                    if img_name.lower().endswith(('.png', '.jpg', '.jpeg')):
                        self.images.append(os.path.join(class_dir, img_name))
                        self.labels.append(self.class_to_idx[class_name])
        
        print(f"Dataset loaded: {len(self.images)} images, {len(self.classes)} classes")
        print(f"Classes: {self.classes}")
    
    def __len__(self):
        return len(self.images)
    
    def __getitem__(self, idx):
        img_path = self.images[idx]
        label = self.labels[idx]
        image = Image.open(img_path).convert('RGB')
        
        if self.transform:
            image = self.transform(image)
        
        return image, label

# ====================
# 3. DATA TRANSFORMS
# ====================
dataset_path = "/kaggle/input/cse-task-475-ssl3/Microscopic Image Dataset of Plant-Parasitic Nematodes"

# MoCo transforms
class TwoCropsTransform:
    def __init__(self, transform):
        self.transform = transform
    
    def __call__(self, x):
        return [self.transform(x), self.transform(x)]

train_transform = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.2, 1.0)),
    transforms.RandomApply([transforms.ColorJitter(0.4, 0.4, 0.4, 0.1)], p=0.8),
    transforms.RandomGrayscale(p=0.2),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

val_transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load dataset
full_dataset = NematodeDataset(dataset_path, transform=None)

# Split dataset
def split_dataset(dataset, train_ratio=0.7, val_ratio=0.15, test_ratio=0.15):
    total_size = len(dataset)
    train_size = int(train_ratio * total_size)
    val_size = int(val_ratio * total_size)
    test_size = total_size - train_size - val_size
    
    train_dataset, val_dataset, test_dataset = random_split(
        dataset, [train_size, val_size, test_size]
    )
    
    return train_dataset, val_dataset, test_dataset

train_dataset, val_dataset, test_dataset = split_dataset(full_dataset)

print(f"Train: {len(train_dataset)}, Val: {len(val_dataset)}, Test: {len(test_dataset)}")

# ====================
# 4. SIMPLE VISION TRANSFORMER (Simplified)
# ====================
class SimpleViT(nn.Module):
    def __init__(self, num_classes=384):
        super().__init__()
        # Simple CNN backbone
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=2, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, stride=2, padding=1)
        self.bn3 = nn.BatchNorm2d(256)
        self.conv4 = nn.Conv2d(256, 512, kernel_size=3, stride=2, padding=1)
        self.bn4 = nn.BatchNorm2d(512)
        
        self.pool = nn.AdaptiveAvgPool2d((1, 1))
        self.relu = nn.ReLU(inplace=True)
        
        # Projection head
        self.fc = nn.Linear(512, num_classes)
        
    def forward(self, x):
        # Check input dimensions
        if x.dim() == 3:
            x = x.unsqueeze(0)  # Add batch dimension if needed
            
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        x = self.relu(self.bn3(self.conv3(x)))
        x = self.relu(self.bn4(self.conv4(x)))
        
        x = self.pool(x)
        x = torch.flatten(x, 1)
        x = self.fc(x)
        
        return x

# ====================
# 5. SIMPLE MOCO IMPLEMENTATION
# ====================
class SimpleMoCo(nn.Module):
    def __init__(self, dim=384, K=4096, m=0.999, T=0.07):
        super().__init__()
        self.K = K
        self.m = m
        self.T = T
        
        # Create encoder
        self.encoder_q = SimpleViT(num_classes=dim)
        self.encoder_k = SimpleViT(num_classes=dim)
        
        # Copy weights
        for param_q, param_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            param_k.data.copy_(param_q.data)
            param_k.requires_grad = False
        
        # Queue
        self.register_buffer("queue", torch.randn(dim, K))
        self.queue = nn.functional.normalize(self.queue, dim=0)
        self.register_buffer("queue_ptr", torch.zeros(1, dtype=torch.long))
    
    @torch.no_grad()
    def _momentum_update_key_encoder(self):
        for param_q, param_k in zip(self.encoder_q.parameters(), self.encoder_k.parameters()):
            param_k.data = param_k.data * self.m + param_q.data * (1. - self.m)
    
    @torch.no_grad()
    def _dequeue_and_enqueue(self, keys):
        batch_size = keys.shape[0]
        ptr = int(self.queue_ptr)
        
        if ptr + batch_size <= self.K:
            self.queue[:, ptr:ptr + batch_size] = keys.T
        else:
            end_size = self.K - ptr
            self.queue[:, ptr:] = keys.T[:, :end_size]
            self.queue[:, :batch_size - end_size] = keys.T[:, end_size:]
        
        ptr = (ptr + batch_size) % self.K
        self.queue_ptr[0] = ptr
    
    def forward(self, im_q, im_k):
        # Query features
        q = self.encoder_q(im_q)
        q = nn.functional.normalize(q, dim=1)
        
        # Key features
        with torch.no_grad():
            self._momentum_update_key_encoder()
            k = self.encoder_k(im_k)
            k = nn.functional.normalize(k, dim=1)
        
        # Compute similarity
        l_pos = torch.einsum('nc,nc->n', [q, k]).unsqueeze(-1)
        l_neg = torch.einsum('nc,ck->nk', [q, self.queue.clone().detach()])
        
        logits = torch.cat([l_pos, l_neg], dim=1) / self.T
        labels = torch.zeros(logits.shape[0], dtype=torch.long).to(device)
        
        self._dequeue_and_enqueue(k)
        
        return logits, labels
    
    def encode(self, x):
        return self.encoder_q(x)

# ====================
# 6. TRAINING SETUP
# ====================
# Create model
moco = SimpleMoCo(dim=384, K=4096, m=0.999, T=0.07)
moco = moco.to(device)

print(f"Trainable parameters: {sum(p.numel() for p in moco.parameters() if p.requires_grad):,}")

# Optimizer
optimizer = optim.AdamW(moco.parameters(), lr=3e-4, weight_decay=0.05)
scheduler = CosineAnnealingLR(optimizer, T_max=50, eta_min=1e-6)
criterion = nn.CrossEntropyLoss()

# ====================
# 7. DATA AUGMENTATION FOR MOCO
# ====================
# Apply transforms
train_dataset.dataset.transform = TwoCropsTransform(train_transform)
val_dataset.dataset.transform = val_transform
test_dataset.dataset.transform = val_transform

# Dataloaders
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

# ====================
# 8. TRAINING FUNCTION
# ====================
def train_epoch(model, dataloader, optimizer, criterion):
    model.train()
    total_loss = 0
    
    for batch_idx, (images, _) in enumerate(tqdm(dataloader, desc="Training")):
        # Get two views for MoCo
        im_q = images[0].to(device)
        im_k = images[1].to(device)
        
        logits, labels = model(im_q, im_k)
        loss = criterion(logits, labels)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        total_loss += loss.item()
    
    return total_loss / len(dataloader)

# ====================
# 9. 50 EPOCHS SSL PRETRAINING
# ====================
epochs = 50
train_losses = []
val_losses = []
train_times = []

print("Starting MoCo SSL pretraining for 50 epochs...")
for epoch in range(epochs):
    start_time = time.time()
    
    # Train
    train_loss = train_epoch(moco, train_loader, optimizer, criterion)
    train_losses.append(train_loss)
    
    # Validate
    moco.eval()
    val_loss = 0
    with torch.no_grad():
        for images, _ in val_loader:
            images = images.to(device)
            # For validation, use same image as both views
            logits, labels = moco(images, images)
            val_loss += criterion(logits, labels).item()
    val_loss /= len(val_loader)
    val_losses.append(val_loss)
    
    # Update scheduler
    scheduler.step()
    
    epoch_time = time.time() - start_time
    train_times.append(epoch_time)
    
    print(f"Epoch {epoch+1}/{epochs}: Loss={train_loss:.4f}, Val={val_loss:.4f}, Time={epoch_time:.2f}s")

print("SSL pretraining completed!")

# ====================
# 10. SAVE MODEL
# ====================
torch.save({
    'model_state_dict': moco.state_dict(),
    'train_losses': train_losses,
    'val_losses': val_losses
}, 'moco_ssl_50ep.pth')
print("Model saved to moco_ssl_50ep.pth")

# ====================
# 11. PLOT TRAINING CURVES
# ====================
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(train_losses, label='Train Loss', linewidth=2)
plt.plot(val_losses, label='Val Loss', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('MoCo Training Loss')
plt.legend()
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(train_times, marker='o')
plt.xlabel('Epoch')
plt.ylabel('Time (seconds)')
plt.title('Training Time per Epoch')
plt.grid(True)

plt.subplot(1, 3, 3)
cumulative_time = np.cumsum(train_times) / 60
plt.plot(cumulative_time, marker='s')
plt.xlabel('Epoch')
plt.ylabel('Cumulative Time (minutes)')
plt.title('Total Training Time')
plt.grid(True)

plt.tight_layout()
plt.show()

# Training statistics
print(f"\nTraining Statistics:")
print(f"• Total training time: {sum(train_times)/60:.2f} minutes")
print(f"• Average epoch time: {np.mean(train_times):.2f} seconds")
print(f"• Final train loss: {train_losses[-1]:.4f}")
print(f"• Final validation loss: {val_losses[-1]:.4f}")
```

```python
# ====================
# 12. EXTRACT EMBEDDINGS
# ====================
def extract_embeddings(model, dataloader):
    model.eval()
    embeddings = []
    labels = []
    
    with torch.no_grad():
        for images, targets in tqdm(dataloader, desc="Extracting"):
            images = images.to(device)
            features = model.encode(images)
            embeddings.append(features.cpu())
            labels.append(targets)
    
    return torch.cat(embeddings, dim=0).numpy(), torch.cat(labels, dim=0).numpy()

print("\nExtracting embeddings...")
X_train, y_train = extract_embeddings(moco, train_loader)
X_val, y_val = extract_embeddings(moco, val_loader)
X_test, y_test = extract_embeddings(moco, test_loader)

print(f"Shapes - Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")

# Check unique classes
unique_train = np.unique(y_train)
unique_val = np.unique(y_val)
unique_test = np.unique(y_test)

print(f"\nUnique classes in Train: {unique_train}")
print(f"Unique classes in Val: {unique_val}")
print(f"Unique classes in Test: {unique_test}")

# Find all classes present
all_classes = sorted(list(set(unique_train) | set(unique_val) | set(unique_test)))
print(f"\nAll unique classes across splits: {all_classes}")
print(f"Total classes: {len(all_classes)}")

# Get class names for existing classes
existing_classes = [full_dataset.classes[i] for i in all_classes]
print(f"\nExisting class names: {existing_classes}")

# Class distribution
print("\nClass Distribution:")
for i, cls in enumerate(all_classes):
    train_count = (y_train == cls).sum()
    val_count = (y_val == cls).sum()
    test_count = (y_test == cls).sum()
    print(f"  Class {cls} ({existing_classes[i]}): Train={train_count}, Val={val_count}, Test={test_count}")

# ====================
# 13. LINEAR PROBE
# ====================
from sklearn.linear_model import LogisticRegression

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Train linear classifier
print("\nTraining Linear Probe...")
linear_clf = LogisticRegression(max_iter=1000, random_state=42, C=1.0)
linear_clf.fit(X_train_scaled, y_train)

# Evaluate
train_acc = linear_clf.score(X_train_scaled, y_train) * 100
val_acc = linear_clf.score(X_val_scaled, y_val) * 100
test_acc = linear_clf.score(X_test_scaled, y_test) * 100

print(f"Linear Probe Results:")
print(f"  Train: {train_acc:.2f}%")
print(f"  Val: {val_acc:.2f}%")
print(f"  Test: {test_acc:.2f}%")

# Classification report with proper labels
y_pred = linear_clf.predict(X_test_scaled)

# Get labels for existing classes
label_names = [full_dataset.classes[i] for i in all_classes]

print("\nClassification Report:")
print(classification_report(y_test, y_pred, labels=all_classes, target_names=label_names))

# ====================
# 14. K-NN EVALUATION
# ====================
print("\nK-NN Evaluation...")
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)

knn_train = knn.score(X_train_scaled, y_train) * 100
knn_val = knn.score(X_val_scaled, y_val) * 100
knn_test = knn.score(X_test_scaled, y_test) * 100

print(f"K-NN (k=5) Results:")
print(f"  Train: {knn_train:.2f}%")
print(f"  Val: {knn_val:.2f}%")
print(f"  Test: {knn_test:.2f}%")

# ====================
# 15. OTHER CLASSIFIERS
# ====================
print("\nTraining other classifiers...")

classifiers = {
    'MLP': MLPClassifier(hidden_layer_sizes=(256, 128), max_iter=500, random_state=42),
    'SVM': SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42),
    'Decision Tree': DecisionTreeClassifier(max_depth=10, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
}

results = {}
for name, clf in classifiers.items():
    print(f"\nTraining {name}...")
    clf.fit(X_train_scaled, y_train)
    
    train_acc = clf.score(X_train_scaled, y_train) * 100
    val_acc = clf.score(X_val_scaled, y_val) * 100
    test_acc = clf.score(X_test_scaled, y_test) * 100
    
    results[name] = {
        'train': train_acc,
        'val': val_acc,
        'test': test_acc
    }
    
    print(f"  {name}: Train={train_acc:.2f}%, Val={val_acc:.2f}%, Test={test_acc:.2f}%")

# ====================
# 16. LABEL EFFICIENCY (COMPLETELY FIXED)
# ====================
print("\nLabel Efficiency Analysis...")

# Calculate minimum viable subset sizes
min_samples_per_class = np.min([(y_train == cls).sum() for cls in all_classes])
print(f"Minimum samples per class in training: {min_samples_per_class}")

# Adjust subset sizes based on available data
subset_sizes = []
if len(X_train) * 0.05 >= len(all_classes):  # At least 1 sample per class
    subset_sizes.append(0.05)
if len(X_train) * 0.1 >= len(all_classes):   # At least 1 sample per class
    subset_sizes.append(0.1)
if len(X_train) * 0.2 >= len(all_classes):   # At least 1 sample per class
    subset_sizes.append(0.2)

# Add larger subsets
subset_sizes.extend([0.3, 0.5, 0.7, 1.0])

print(f"Using subset sizes: {subset_sizes}")

label_efficiency = []

for size in subset_sizes:
    if size < 1.0:
        # Calculate exact number of samples
        n_samples = int(len(X_train) * size)
        
        # Ensure we have at least 1 sample per class
        if n_samples < len(all_classes):
            print(f"  Skipping {size*100:.1f}%: Only {n_samples} samples for {len(all_classes)} classes")
            label_efficiency.append(0)  # Placeholder
            continue
            
        # Manual stratified sampling
        from collections import defaultdict
        class_indices = defaultdict(list)
        for idx, label in enumerate(y_train):
            class_indices[label].append(idx)
        
        subset_indices = []
        samples_per_class = max(1, n_samples // len(all_classes))
        
        for cls in all_classes:
            if cls in class_indices:
                available = class_indices[cls]
                n_to_take = min(samples_per_class, len(available))
                subset_indices.extend(np.random.choice(available, n_to_take, replace=False))
            else:
                print(f"  Warning: Class {cls} not in training set")
        
        # If we need more samples, add randomly
        if len(subset_indices) < n_samples:
            all_indices = list(range(len(X_train)))
            remaining_indices = [i for i in all_indices if i not in subset_indices]
            additional_needed = n_samples - len(subset_indices)
            if additional_needed > 0 and remaining_indices:
                subset_indices.extend(np.random.choice(remaining_indices, additional_needed, replace=False))
        
        X_sub = X_train_scaled[subset_indices]
        y_sub = y_train[subset_indices]
        
        print(f"  Subset {size*100:.1f}%: {len(subset_indices)} samples, {len(np.unique(y_sub))} classes")
    else:
        X_sub = X_train_scaled
        y_sub = y_train
    
    # Train classifier
    clf = LogisticRegression(max_iter=500, random_state=42)
    clf.fit(X_sub, y_sub)
    
    test_acc = clf.score(X_test_scaled, y_test) * 100
    label_efficiency.append(test_acc)
    
    print(f"  {size*100:.1f}% data: Test Acc = {test_acc:.2f}%")

# ====================
# 17. FULL FINE-TUNING (SIMPLIFIED)
# ====================
print("\nFull Fine-tuning Comparison...")

# Simple CNN for fine-tuning (same as SSL backbone)
class FineTuneCNN(nn.Module):
    def __init__(self, num_classes=len(all_classes)):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=2, padding=1)
        self.bn1 = nn.BatchNorm2d(64)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, stride=2, padding=1)
        self.bn2 = nn.BatchNorm2d(128)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, stride=2, padding=1)
        self.bn3 = nn.BatchNorm2d(256)
        self.conv4 = nn.Conv2d(256, 512, kernel_size=3, stride=2, padding=1)
        self.bn4 = nn.BatchNorm2d(512)
        
        self.pool = nn.AdaptiveAvgPool2d((1, 1))
        self.relu = nn.ReLU(inplace=True)
        self.dropout = nn.Dropout(0.5)
        self.fc = nn.Linear(512, num_classes)
        
    def forward(self, x):
        x = self.relu(self.bn1(self.conv1(x)))
        x = self.relu(self.bn2(self.conv2(x)))
        x = self.relu(self.bn3(self.conv3(x)))
        x = self.relu(self.bn4(self.conv4(x)))
        
        x = self.pool(x)
        x = torch.flatten(x, 1)
        x = self.dropout(x)
        x = self.fc(x)
        
        return x

# Create model for fine-tuning
num_classes_actual = len(all_classes)
ft_model = FineTuneCNN(num_classes=num_classes_actual)
ft_model = ft_model.to(device)

print(f"Fine-tuning model with {num_classes_actual} classes")

# Use the same train/val/test splits from SSL (already loaded)
# Just update transforms for fine-tuning
ft_transform = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Update transforms in existing datasets
train_dataset.dataset.transform = ft_transform
val_dataset.dataset.transform = val_transform

# Recreate loaders with new transforms
ft_train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
ft_val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

# Fine-tuning setup
ft_criterion = nn.CrossEntropyLoss()
ft_optimizer = optim.AdamW(ft_model.parameters(), lr=1e-4, weight_decay=0.05)
ft_scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(ft_optimizer, T_max=30)

# Training loop
ft_epochs = 30
best_val_acc = 0
best_model = None
ft_train_acc_history = []
ft_val_acc_history = []

print("\nStarting fine-tuning...")
for epoch in range(ft_epochs):
    ft_model.train()
    train_correct = 0
    train_total = 0
    
    for images, labels in ft_train_loader:
        images, labels = images.to(device), labels.to(device)
        
        ft_optimizer.zero_grad()
        outputs = ft_model(images)
        loss = ft_criterion(outputs, labels)
        loss.backward()
        ft_optimizer.step()
        
        _, predicted = torch.max(outputs, 1)
        train_total += labels.size(0)
        train_correct += (predicted == labels).sum().item()
    
    train_acc = 100. * train_correct / train_total
    ft_train_acc_history.append(train_acc)
    
    # Validation
    ft_model.eval()
    val_correct = 0
    val_total = 0
    
    with torch.no_grad():
        for images, labels in ft_val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = ft_model(images)
            _, predicted = torch.max(outputs, 1)
            val_total += labels.size(0)
            val_correct += (predicted == labels).sum().item()
    
    val_acc = 100. * val_correct / val_total
    ft_val_acc_history.append(val_acc)
    
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_model = copy.deepcopy(ft_model.state_dict())
    
    ft_scheduler.step()
    
    if (epoch + 1) % 5 == 0:
        print(f"  Epoch {epoch+1}: Train={train_acc:.2f}%, Val={val_acc:.2f}%")

# Load best model
ft_model.load_state_dict(best_model)

# Test evaluation
ft_model.eval()
test_correct = 0
test_total = 0
ft_predictions = []
ft_true_labels = []

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = ft_model(images)
        _, predicted = torch.max(outputs, 1)
        test_total += labels.size(0)
        test_correct += (predicted == labels).sum().item()
        
        ft_predictions.extend(predicted.cpu().numpy())
        ft_true_labels.extend(labels.cpu().numpy())

ft_test_acc = 100. * test_correct / test_total
print(f"\nFine-tuned Model Test Accuracy: {ft_test_acc:.2f}%")

# Convert to numpy arrays
ft_predictions = np.array(ft_predictions)
ft_true_labels = np.array(ft_true_labels)

print("\nFine-tuned Classification Report:")
print(classification_report(ft_true_labels, ft_predictions, labels=all_classes, target_names=label_names))
```

```python
# ====================
# 17.5. EMBEDDING VISUALIZATION AND CLUSTERING METRICS
# ====================
print("\n" + "="*60)
print("EMBEDDING VISUALIZATION AND CLUSTERING ANALYSIS")
print("="*60)

from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import umap.umap_ as umap  # Fixed UMAP import
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score, davies_bouldin_score
import seaborn as sns

# Prepare data for visualization
print("\nPreparing data for visualization...")

# Use all embeddings for better visualization
X_all = np.vstack([X_train_scaled, X_val_scaled, X_test_scaled])
y_all = np.concatenate([y_train, y_val, y_test])
split_labels = np.concatenate([
    np.full(len(X_train_scaled), 'Train'),
    np.full(len(X_val_scaled), 'Val'),
    np.full(len(X_test_scaled), 'Test')
])

print(f"Total embeddings for visualization: {len(X_all)}")
print(f"Class distribution: {np.bincount(y_all)}")

# ====================
# 1. PCA Visualization (2D and 3D)
# ====================
print("\n1. Performing PCA...")

# 2D PCA
pca_2d = PCA(n_components=2, random_state=42)
X_pca_2d = pca_2d.fit_transform(X_all)

print(f"PCA explained variance ratio: {pca_2d.explained_variance_ratio_}")
print(f"Total explained variance: {pca_2d.explained_variance_ratio_.sum():.3f}")

# 3D PCA
pca_3d = PCA(n_components=3, random_state=42)
X_pca_3d = pca_3d.fit_transform(X_all)

# ====================
# 2. t-SNE Visualization
# ====================
print("\n2. Performing t-SNE...")

# Determine optimal perplexity
perplexity_value = min(30, len(X_all) - 1)
print(f"Using perplexity: {perplexity_value}")

X_tsne_2d = None
silhouette_tsne = None
ch_tsne = None
db_tsne = None

try:
    tsne_2d = TSNE(n_components=2, random_state=42, perplexity=perplexity_value, 
                   n_iter=1000, learning_rate=200)
    X_tsne_2d = tsne_2d.fit_transform(X_all)
    print("t-SNE 2D completed successfully")
    
    # Calculate metrics for t-SNE
    silhouette_tsne = silhouette_score(X_tsne_2d, y_all)
    ch_tsne = calinski_harabasz_score(X_tsne_2d, y_all)
    db_tsne = davies_bouldin_score(X_tsne_2d, y_all)
    
except Exception as e:
    print(f"t-SNE 2D failed: {e}")
    # Use PCA as fallback
    X_tsne_2d = X_pca_2d.copy()
    print("Using PCA as fallback for t-SNE")

# ====================
# 3. UMAP Visualization
# ====================
print("\n3. Performing UMAP...")

X_umap_2d = None
silhouette_umap = None
ch_umap = None
db_umap = None

try:
    umap_reducer = umap.UMAP(n_components=2, random_state=42, n_neighbors=15, min_dist=0.1)
    X_umap_2d = umap_reducer.fit_transform(X_all)
    print("UMAP 2D completed successfully")
    
    # Calculate metrics for UMAP
    silhouette_umap = silhouette_score(X_umap_2d, y_all)
    ch_umap = calinski_harabasz_score(X_umap_2d, y_all)
    db_umap = davies_bouldin_score(X_umap_2d, y_all)
    
except Exception as e:
    print(f"UMAP 2D failed: {e}")
    # Use PCA as fallback
    X_umap_2d = X_pca_2d.copy()
    print("Using PCA as fallback for UMAP")

# ====================
# 4. CLUSTERING METRICS
# ====================
print("\n4. Calculating clustering metrics...")

# Silhouette Score (higher is better, range: -1 to 1)
silhouette_pca = silhouette_score(X_pca_2d, y_all)

# Calinski-Harabasz Index (higher is better)
ch_pca = calinski_harabasz_score(X_pca_2d, y_all)

# Davies-Bouldin Index (lower is better)
db_pca = davies_bouldin_score(X_pca_2d, y_all)

print("\nClustering Metrics:")
print("-" * 60)
print(f"{'Metric':<20} {'PCA':<10} {'t-SNE':<10} {'UMAP':<10}")
print("-" * 60)

# Format t-SNE values
tsne_silhouette_str = f"{silhouette_tsne:.4f}" if silhouette_tsne is not None else "N/A"
tsne_ch_str = f"{ch_tsne:.2f}" if ch_tsne is not None else "N/A"
tsne_db_str = f"{db_tsne:.4f}" if db_tsne is not None else "N/A"

# Format UMAP values
umap_silhouette_str = f"{silhouette_umap:.4f}" if silhouette_umap is not None else "N/A"
umap_ch_str = f"{ch_umap:.2f}" if ch_umap is not None else "N/A"
umap_db_str = f"{db_umap:.4f}" if db_umap is not None else "N/A"

print(f"{'Silhouette Score':<20} {silhouette_pca:<10.4f} {tsne_silhouette_str:<10} {umap_silhouette_str:<10}")
print(f"{'Calinski-Harabasz':<20} {ch_pca:<10.2f} {tsne_ch_str:<10} {umap_ch_str:<10}")
print(f"{'Davies-Bouldin':<20} {db_pca:<10.4f} {tsne_db_str:<10} {umap_db_str:<10}")
print("-" * 60)

# Interpretation
print("\nInterpretation:")
print("• Silhouette Score: Measures how similar objects are to their own cluster (cohesion)")
print("  compared to other clusters (separation). Range: -1 (poor) to 1 (excellent)")
print("• Calinski-Harabasz: Ratio of between-cluster dispersion to within-cluster dispersion.")
print("  Higher values indicate better defined clusters.")
print("• Davies-Bouldin: Average similarity between each cluster and its most similar one.")
print("  Lower values indicate better separation.")

# ====================
# 5. VISUALIZATIONS
# ====================
print("\n5. Creating visualizations...")

# Create a large figure with all visualizations
fig = plt.figure(figsize=(20, 15))

# Color palette for classes
palette = sns.color_palette("husl", len(all_classes))

# 1. PCA 2D Plot
ax1 = plt.subplot(3, 4, 1)
for i, cls in enumerate(all_classes):
    mask = y_all == cls
    ax1.scatter(X_pca_2d[mask, 0], X_pca_2d[mask, 1], 
                color=palette[i], alpha=0.6, s=20, label=label_names[i])
ax1.set_xlabel('PC1')
ax1.set_ylabel('PC2')
ax1.set_title(f'PCA 2D (Explained Variance: {pca_2d.explained_variance_ratio_.sum():.2%})')
ax1.grid(True, alpha=0.3)

# 2. PCA 3D Plot
ax2 = plt.subplot(3, 4, 2, projection='3d')
for i, cls in enumerate(all_classes):
    mask = y_all == cls
    ax2.scatter(X_pca_3d[mask, 0], X_pca_3d[mask, 1], X_pca_3d[mask, 2],
                color=palette[i], alpha=0.6, s=20, label=label_names[i])
ax2.set_xlabel('PC1')
ax2.set_ylabel('PC2')
ax2.set_zlabel('PC3')
ax2.set_title('PCA 3D Visualization')

# 3. t-SNE 2D Plot
ax3 = plt.subplot(3, 4, 3)
if X_tsne_2d is not None:
    for i, cls in enumerate(all_classes):
        mask = y_all == cls
        ax3.scatter(X_tsne_2d[mask, 0], X_tsne_2d[mask, 1], 
                    color=palette[i], alpha=0.6, s=20, label=label_names[i])
    
    # Format title based on whether we have silhouette score
    tsne_title = 't-SNE 2D'
    if silhouette_tsne is not None:
        tsne_title = f't-SNE 2D (Silhouette: {silhouette_tsne:.3f})'
    else:
        tsne_title = 't-SNE 2D (Using PCA)'
    
    ax3.set_title(tsne_title)
else:
    ax3.text(0.5, 0.5, 't-SNE failed\nUsing PCA', 
             ha='center', va='center', transform=ax3.transAxes)
    ax3.set_title('t-SNE 2D (Failed)')

ax3.set_xlabel('t-SNE 1')
ax3.set_ylabel('t-SNE 2')
ax3.grid(True, alpha=0.3)

# 4. UMAP 2D Plot
ax4 = plt.subplot(3, 4, 4)
if X_umap_2d is not None:
    for i, cls in enumerate(all_classes):
        mask = y_all == cls
        ax4.scatter(X_umap_2d[mask, 0], X_umap_2d[mask, 1], 
                    color=palette[i], alpha=0.6, s=20, label=label_names[i])
    
    # Format title based on whether we have silhouette score
    umap_title = 'UMAP 2D'
    if silhouette_umap is not None:
        umap_title = f'UMAP 2D (Silhouette: {silhouette_umap:.3f})'
    else:
        umap_title = 'UMAP 2D (Using PCA)'
    
    ax4.set_title(umap_title)
else:
    ax4.text(0.5, 0.5, 'UMAP failed\nUsing PCA', 
             ha='center', va='center', transform=ax4.transAxes)
    ax4.set_title('UMAP 2D (Failed)')

ax4.set_xlabel('UMAP 1')
ax4.set_ylabel('UMAP 2')
ax4.grid(True, alpha=0.3)

# 5. Split-wise visualization (Train/Val/Test)
ax5 = plt.subplot(3, 4, 5)
for split, color in zip(['Train', 'Val', 'Test'], ['blue', 'green', 'red']):
    mask = split_labels == split
    ax5.scatter(X_pca_2d[mask, 0], X_pca_2d[mask, 1], 
                color=color, alpha=0.5, s=15, label=split)
ax5.set_xlabel('PC1')
ax5.set_ylabel('PC2')
ax5.set_title('PCA - Data Split Visualization')
ax5.legend()
ax5.grid(True, alpha=0.3)

# 6. Class distribution bar plot
ax6 = plt.subplot(3, 4, 6)
class_counts = np.bincount(y_all)
ax6.bar(range(len(class_counts)), class_counts, color=palette)
ax6.set_xlabel('Class Index')
ax6.set_ylabel('Number of Samples')
ax6.set_title('Class Distribution')
ax6.grid(True, alpha=0.3, axis='y')

# 7. Silhouette analysis
ax7 = plt.subplot(3, 4, 7)
metrics_data = {
    'PCA': silhouette_pca,
    't-SNE': silhouette_tsne if silhouette_tsne is not None else 0,
    'UMAP': silhouette_umap if silhouette_umap is not None else 0
}
methods = list(metrics_data.keys())
scores = list(metrics_data.values())
colors = ['skyblue', 'lightcoral', 'lightgreen']
bars = ax7.bar(methods, scores, color=colors)
ax7.set_ylabel('Silhouette Score')
ax7.set_title('Clustering Quality Comparison')
ax7.set_ylim([-1, 1])
ax7.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax7.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, score in zip(bars, scores):
    height = bar.get_height()
    ax7.text(bar.get_x() + bar.get_width()/2., height + 0.02,
             f'{score:.3f}', ha='center', va='bottom')

# 8. Explained variance plot
ax8 = plt.subplot(3, 4, 8)
cumulative_variance = np.cumsum(pca_2d.explained_variance_ratio_)
components = range(1, len(cumulative_variance) + 1)
ax8.plot(components, cumulative_variance, 'bo-', linewidth=2)
ax8.set_xlabel('Number of Components')
ax8.set_ylabel('Cumulative Explained Variance')
ax8.set_title('PCA - Cumulative Explained Variance')
ax8.grid(True, alpha=0.3)
ax8.axhline(y=0.95, color='r', linestyle='--', alpha=0.5, label='95% variance')
ax8.legend()

# 9. Pairwise class separation (heatmap)
ax9 = plt.subplot(3, 4, 9)
# Calculate average distance between class centroids
centroids = []
for cls in all_classes:
    mask = y_all == cls
    centroid = X_pca_2d[mask].mean(axis=0)
    centroids.append(centroid)
centroids = np.array(centroids)

# Calculate pairwise distances
from scipy.spatial.distance import cdist
pairwise_dist = cdist(centroids, centroids, 'euclidean')

# Plot heatmap
im = ax9.imshow(pairwise_dist, cmap='YlOrRd', aspect='auto')
ax9.set_xlabel('Class Index')
ax9.set_ylabel('Class Index')
ax9.set_title('Pairwise Class Distances (PCA)')
plt.colorbar(im, ax=ax9)

# 10. Density plot of embeddings
ax10 = plt.subplot(3, 4, 10)
from scipy.stats import gaussian_kde
# Use first two PCA components
xy = np.vstack([X_pca_2d[:, 0], X_pca_2d[:, 1]])
z = gaussian_kde(xy)(xy)
sc = ax10.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=z, s=20, cmap='viridis')
ax10.set_xlabel('PC1')
ax10.set_ylabel('PC2')
ax10.set_title('Density of Embeddings')
plt.colorbar(sc, ax=ax10)

# 11. Class-wise silhouette values
ax11 = plt.subplot(3, 4, 11)
from sklearn.metrics import silhouette_samples
silhouette_vals = silhouette_samples(X_pca_2d, y_all)
class_silhouette = []
for cls in all_classes:
    mask = y_all == cls
    class_silhouette.append(silhouette_vals[mask].mean())

ax11.bar(range(len(class_silhouette)), class_silhouette, color=palette)
ax11.set_xlabel('Class Index')
ax11.set_ylabel('Average Silhouette')
ax11.set_title('Per-class Silhouette Scores')
ax11.grid(True, alpha=0.3, axis='y')
ax11.axhline(y=np.mean(silhouette_vals), color='red', linestyle='--', 
             label=f'Mean: {np.mean(silhouette_vals):.3f}')
ax11.legend()

# 12. Quality metrics radar chart
ax12 = plt.subplot(3, 4, 12, projection='polar')
# Normalize metrics for radar chart
def normalize_metric(metric_values):
    valid_values = [v for v in metric_values if v is not None]
    if not valid_values:
        return [0, 0, 0]
    min_val = min(valid_values)
    max_val = max(valid_values)
    if max_val == min_val:
        return [0.5 if v is not None else 0 for v in metric_values]
    return [(v - min_val) / (max_val - min_val) if v is not None else 0 for v in metric_values]

# Prepare data for radar chart
methods_radar = ['PCA', 't-SNE', 'UMAP']
silhouette_norm = normalize_metric([silhouette_pca, silhouette_tsne, silhouette_umap])
ch_norm = normalize_metric([ch_pca, ch_tsne, ch_umap])
db_norm = [1 - v for v in normalize_metric([db_pca, db_tsne, db_umap])]  # Invert DB (lower is better)

categories = ['Silhouette', 'Calinski\nHarabasz', 'Davies\nBouldin*']
N = len(categories)

angles = [n / float(N) * 2 * np.pi for n in range(N)]
angles += angles[:1]

for idx, method in enumerate(methods_radar):
    values = [silhouette_norm[idx], ch_norm[idx], db_norm[idx]]
    values += values[:1]
    ax12.plot(angles, values, linewidth=2, linestyle='solid', label=method)
    ax12.fill(angles, values, alpha=0.1)

ax12.set_xticks(angles[:-1])
ax12.set_xticklabels(categories)
ax12.set_title('Normalized Clustering Metrics\n(*DB inverted: higher is better)')
ax12.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))

plt.suptitle('Comprehensive Embedding Analysis: SSL with MoCo + ViT', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# ====================
# 6. ADDITIONAL ANALYSIS
# ====================
print("\n6. Additional Analysis...")

# Variance explained by top components
print(f"\nVariance explained by top 5 PCA components:")
for i in range(min(5, len(pca_2d.explained_variance_ratio_))):
    print(f"  Component {i+1}: {pca_2d.explained_variance_ratio_[i]:.3%}")

# Class separability analysis
print(f"\nClass Separability Analysis:")
print("-" * 40)

# Calculate within-class and between-class scatter
n_classes = len(all_classes)
n_features = X_all.shape[1]

# Overall mean
overall_mean = X_all.mean(axis=0)

# Within-class scatter matrix
S_W = np.zeros((n_features, n_features))
for cls in all_classes:
    mask = y_all == cls
    X_class = X_all[mask]
    mean_class = X_class.mean(axis=0)
    
    # Center the class data
    X_class_centered = X_class - mean_class
    
    # Add to within-class scatter
    S_W += X_class_centered.T @ X_class_centered

# Between-class scatter matrix
S_B = np.zeros((n_features, n_features))
for cls in all_classes:
    mask = y_all == cls
    n_class = mask.sum()
    mean_class = X_all[mask].mean(axis=0)
    
    # Difference between class mean and overall mean
    mean_diff = (mean_class - overall_mean).reshape(-1, 1)
    
    # Add to between-class scatter
    S_B += n_class * (mean_diff @ mean_diff.T)

# Calculate Fisher's discriminant ratio (trace criterion)
try:
    J = np.trace(np.linalg.pinv(S_W) @ S_B)
    print(f"Fisher's Discriminant Ratio (J): {J:.4f}")
    print("Higher values indicate better class separability.")
except:
    print("Could not calculate Fisher's Discriminant Ratio (singular matrix)")

# Calculate class compactness (average within-class distance)
class_compactness = []
for cls in all_classes:
    mask = y_all == cls
    X_class = X_all[mask]
    if len(X_class) > 1:
        # Calculate pairwise distances within class
        from scipy.spatial.distance import pdist
        distances = pdist(X_class)
        class_compactness.append(distances.mean())

avg_compactness = np.mean(class_compactness) if class_compactness else 0
print(f"Average within-class distance: {avg_compactness:.4f}")
print("Lower values indicate tighter clusters.")

# Calculate separation between classes
class_separation = []
for i, cls1 in enumerate(all_classes):
    for j, cls2 in enumerate(all_classes):
        if i < j:
            mask1 = y_all == cls1
            mask2 = y_all == cls2
            X1 = X_all[mask1]
            X2 = X_all[mask2]
            
            # Calculate distance between class centroids
            centroid1 = X1.mean(axis=0)
            centroid2 = X2.mean(axis=0)
            distance = np.linalg.norm(centroid1 - centroid2)
            class_separation.append(distance)

avg_separation = np.mean(class_separation) if class_separation else 0
print(f"Average between-class distance: {avg_separation:.4f}")
print("Higher values indicate better separation.")

# Separation-to-compactness ratio
if avg_compactness > 0:
    separation_ratio = avg_separation / avg_compactness
    print(f"Separation-to-compactness ratio: {separation_ratio:.4f}")
    print("Higher ratios indicate better clustering quality.")

# ====================
# 7. SAVE VISUALIZATION RESULTS
# ====================
print("\n7. Saving results...")

# Create results dictionary
visualization_results = {
    'pca_explained_variance': pca_2d.explained_variance_ratio_.tolist(),
    'pca_cumulative_variance': cumulative_variance.tolist(),
    'clustering_metrics': {
        'silhouette': {'pca': float(silhouette_pca), 
                      'tsne': float(silhouette_tsne) if silhouette_tsne is not None else None,
                      'umap': float(silhouette_umap) if silhouette_umap is not None else None},
        'calinski_harabasz': {'pca': float(ch_pca),
                             'tsne': float(ch_tsne) if ch_tsne is not None else None,
                             'umap': float(ch_umap) if ch_umap is not None else None},
        'davies_bouldin': {'pca': float(db_pca),
                          'tsne': float(db_tsne) if db_tsne is not None else None,
                          'umap': float(db_umap) if db_umap is not None else None}
    },
    'class_separability': {
        'fishers_discriminant': float(J) if 'J' in locals() else None,
        'avg_within_class_distance': float(avg_compactness),
        'avg_between_class_distance': float(avg_separation),
        'separation_ratio': float(separation_ratio) if 'separation_ratio' in locals() else None
    },
    'embedding_statistics': {
        'total_samples': len(X_all),
        'num_classes': len(all_classes),
        'embedding_dim': X_all.shape[1],
        'class_distribution': class_counts.tolist()
    }
}

# Save to JSON
import json
with open('embedding_analysis_results.json', 'w') as f:
    json.dump(visualization_results, f, indent=2)

print("✅ Embedding analysis results saved to 'embedding_analysis_results.json'")

# Save the main visualization figure
plt.figure(figsize=(15, 10))
# Recreate the main PCA plot for saving
for i, cls in enumerate(all_classes):
    mask = y_all == cls
    plt.scatter(X_pca_2d[mask, 0], X_pca_2d[mask, 1], 
                color=palette[i], alpha=0.6, s=30, label=label_names[i])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title(f'SSL Embeddings - PCA Visualization\nSilhouette Score: {silhouette_pca:.3f}')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('ssl_embeddings_pca.png', dpi=300, bbox_inches='tight')
plt.close()

print("✅ PCA visualization saved to 'ssl_embeddings_pca.png'")

print("\n" + "="*60)
print("EMBEDDING ANALYSIS COMPLETE!")
print("="*60)
```
