```python
!pip install thop --quiet

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, random_split, Subset
from torchvision import datasets, transforms, models
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc, silhouette_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
import umap
from thop import profile
import time, os, random
from thop import profile

print("THOP imported successfully!")



```

```python
# ===============================================================
# 2. Device Setup
# ===============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# ===============================================================
# 3. Dataset Paths & Transforms
# ===============================================================
DATA_DIR = "/kaggle/input/nematoda/Microscopic Image Dataset of Plant-Parasitic Nematodes"

ssl_transform = transforms.Compose([
    transforms.RandomResizedCrop(224),
    transforms.RandomHorizontalFlip(),
    transforms.RandomApply([transforms.ColorJitter(.4, .4, .4, .1)], p=0.8),
    transforms.RandomGrayscale(p=0.2),
    transforms.GaussianBlur(3),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

eval_transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485,0.456,0.406],[0.229,0.224,0.225])
])

full_dataset = datasets.ImageFolder(DATA_DIR, transform=ssl_transform)
eval_dataset = datasets.ImageFolder(DATA_DIR, transform=eval_transform)

```

```python
# ===============================================================
# 4. Train/Test/Val Split
# ===============================================================
test_ratio = 0.1
test_size = int(len(full_dataset)*test_ratio)
train_size = len(full_dataset)-test_size

ssl_train_dataset, ssl_test_dataset = random_split(full_dataset, [train_size, test_size])
train_dataset, test_dataset = random_split(eval_dataset, [train_size, test_size])

val_size = int(0.1 * len(train_dataset))
train_size_final = len(train_dataset) - val_size
train_dataset, val_dataset = random_split(train_dataset, [train_size_final, val_size])

print(f"SSL Train: {len(ssl_train_dataset)}, SSL Test: {len(ssl_test_dataset)}")
print(f"Train: {len(train_dataset)}, Val: {len(val_dataset)}, Test: {len(test_dataset)}")
```

```python
# ===============================================================
# 5. DataLoaders
# ===============================================================
batch_size = 32
ssl_loader = DataLoader(ssl_train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
test_loader = DataLoader(test_dataset, batch_size=batch_size)
```

```python
import torch
import torch.nn as nn
import copy

# ------------------------
# MLPs for BYOL
# ------------------------
class projection_MLP(nn.Module):
    def __init__(self, in_dim=2048, hidden_dim=4096, out_dim=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_dim, out_dim)
        )
    def forward(self, x):
        return self.net(x)

class prediction_MLP(nn.Module):
    def __init__(self, in_dim=256, hidden_dim=4096, out_dim=256):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(in_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(inplace=True),
            nn.Linear(hidden_dim, out_dim)
        )
    def forward(self, x):
        return self.net(x)

# ------------------------
# BYOL Class
# ------------------------
class BYOL(nn.Module):
    def __init__(self, base_encoder):
        super().__init__()

        # -------------------------------
        # Online network
        # -------------------------------
        self.online_encoder = base_encoder
        self.online_projector = projection_MLP()
        self.online_predictor = prediction_MLP()

        # -------------------------------
        # Target network (deep copy = safe)
        # -------------------------------
        self.target_encoder = copy.deepcopy(base_encoder)
        self.target_projector = copy.deepcopy(self.online_projector)

        # Freeze target net
        for p in self.target_encoder.parameters():
            p.requires_grad = False
        for p in self.target_projector.parameters():
            p.requires_grad = False

    @torch.no_grad()
    def _update_target_network(self, momentum=0.996):
        # Encoder update
        for p_o, p_t in zip(self.online_encoder.parameters(), 
                            self.target_encoder.parameters()):
            p_t.data = momentum * p_t.data + (1 - momentum) * p_o.data
        # Projector update
        for p_o, p_t in zip(self.online_projector.parameters(), 
                            self.target_projector.parameters()):
            p_t.data = momentum * p_t.data + (1 - momentum) * p_o.data

    def forward(self, x1, x2):
        # Online network
        o1 = self.online_predictor(self.online_projector(self.online_encoder(x1)))
        o2 = self.online_predictor(self.online_projector(self.online_encoder(x2)))

        # Target network (no gradient)
        with torch.no_grad():
            t1 = self.target_projector(self.target_encoder(x1))
            t2 = self.target_projector(self.target_encoder(x2))

        return o1, t2.detach(), o2, t1.detach()
```

```python
# ===============================================================
# 7. Load ResNet50 backbone & Initialize BYOL
# ===============================================================
import copy
base = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V1)
base.fc = nn.Identity()
model = BYOL(base).to(device)
optimizer = optim.Adam(model.parameters(), lr=3e-4)


```

```python
import torch.nn.functional as F

# ===============================================================
# BYOL Loss Function
# ===============================================================
def byol_loss(p, z):
    """
    Computes BYOL loss between online predictor p and target projection z
    p: online predictor output (batch_size x 256)
    z: target projection output (batch_size x 256)
    """
    p = F.normalize(p, dim=1)  # normalize along feature dim
    z = F.normalize(z, dim=1)
    return 2 - 2 * (p * z).sum(dim=1)  # cosine similarity loss
```

```python
# ===============================================================
# 8. SSL Pretraining Loop
# ===============================================================
epochs_ssl = 50
ssl_losses = []

print("\nğŸ”¥ Starting BYOL SSL Pretraining...\n")
for epoch in range(epochs_ssl):
    model.train()
    total_loss = 0
    for (x1,_), (x2,_) in zip(ssl_loader, ssl_loader):
        x1, x2 = x1.to(device), x2.to(device)
        p1, t2, p2, t1 = model(x1,x2)
        loss = byol_loss(p1,t2).mean() + byol_loss(p2,t1).mean()
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        model._update_target_network()
        total_loss += loss.item()
    ssl_losses.append(total_loss/len(ssl_loader))
    print(f"Epoch {epoch+1}/{epochs_ssl} | SSL Loss: {ssl_losses[-1]:.4f}")

torch.save(model.online_encoder.state_dict(), "byol_resnet50_encoder.pth")

plt.plot(ssl_losses)
plt.title("BYOL SSL Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
# ===============================================================
# 9. Downstream Linear Probe
# ===============================================================
# Use BYOL encoder
encoder = models.resnet50(weights=None)
encoder.fc = nn.Identity()  # output = 2048 features
encoder.load_state_dict(torch.load("byol_resnet50_encoder.pth"), strict=False)
encoder.to(device)
encoder.eval()

for p in encoder.parameters():
    p.requires_grad=False

num_classes = len(eval_dataset.classes)
linear_head = nn.Linear(2048,num_classes).to(device)
optimizer = optim.Adam(linear_head.parameters(), lr=1e-3)
criterion = nn.CrossEntropyLoss()

epochs_lp = 50
train_accs, val_accs = [], []
train_losses, val_losses = [], []

for epoch in range(epochs_lp):
    linear_head.train()
    total, correct, running_loss = 0,0,0
    for imgs,labels in train_loader:
        imgs,labels = imgs.to(device), labels.to(device)
        with torch.no_grad():
            feats = encoder(imgs)
        outputs = linear_head(feats)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss += loss.item()*imgs.size(0)
        _, preds = torch.max(outputs,1)
        correct += (preds==labels).sum().item()
        total += labels.size(0)
    train_accs.append(correct/total)
    train_losses.append(running_loss/total)

    # validation
    linear_head.eval()
    total, correct, val_loss =0,0,0
    with torch.no_grad():
        for imgs,labels in val_loader:
            imgs,labels = imgs.to(device), labels.to(device)
            feats = encoder(imgs)
            outputs = linear_head(feats)
            loss = criterion(outputs,labels)
            val_loss += loss.item()*imgs.size(0)
            _,preds = torch.max(outputs,1)
            correct += (preds==labels).sum().item()
            total += labels.size(0)
    val_accs.append(correct/total)
    val_losses.append(val_loss/total)
    print(f"Epoch {epoch+1}/{epochs_lp} | Train Acc: {train_accs[-1]:.4f} | Val Acc: {val_accs[-1]:.4f}")

torch.save(linear_head.state_dict(),"byol_linear_probe.pth")

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(train_losses,label="Train Loss")
plt.plot(val_losses,label="Val Loss")
plt.title("Linear Probe Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.subplot(1,2,2)
plt.plot(train_accs,label="Train Acc")
plt.plot(val_accs,label="Val Acc")
plt.title("Linear Probe Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
# ===============================================================
# 10. Downstream Test Evaluation (Fixed)
# ===============================================================
linear_head.eval()
all_preds, all_labels, all_probs = [], [], []

with torch.no_grad():
    for imgs, labels in test_loader:
        imgs = imgs.to(device)
        feats = encoder(imgs)
        outputs = linear_head(feats)
        probs = torch.softmax(outputs, dim=1)
        _, preds = torch.max(outputs, 1)
        
        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.numpy())
        all_probs.extend(probs.cpu().numpy())

all_labels = np.array(all_labels)
all_preds = np.array(all_preds)
all_probs = np.array(all_probs)

# Fix target_names mismatch
num_classes = len(np.unique(all_labels))
target_names = eval_dataset.classes[:num_classes]  # slice to match

print(classification_report(all_labels, all_preds, labels=list(range(num_classes)), target_names=target_names))

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds, labels=list(range(num_classes)))
plt.figure(figsize=(8,6))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.colorbar()
plt.xticks(ticks=range(num_classes), labels=target_names, rotation=45)
plt.yticks(ticks=range(num_classes), labels=target_names)
plt.show()

# ROC Curves per class
plt.figure(figsize=(10,8))
for i, cls in enumerate(target_names):
    y_true = (all_labels == i).astype(int)
    y_score = all_probs[:, i]
    fpr, tpr, _ = roc_curve(y_true, y_score)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f"{cls} AUC={roc_auc:.3f}")
plt.plot([0,1], [0,1], 'k--')  # random baseline
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves")
plt.legend()
plt.show()
```

```python
# ===============================================================
# 15. Label-Efficiency Curves (Linear Probe on Subsets)
# ===============================================================

# Make sure encoder outputs features, not logits
encoder.eval()                   # set to eval mode
encoder.fc = nn.Identity()       # output = 2048 features

label_ratios = [0.01, 0.05, 0.10, 0.25, 0.50]
subset_train_accs = []
subset_val_accs = []

for ratio in label_ratios:
    subset_size = max(1, int(len(train_dataset)*ratio))
    subset_indices = random.sample(range(len(train_dataset)), subset_size)
    subset = Subset(train_dataset, subset_indices)
    subset_loader = DataLoader(subset, batch_size=batch_size, shuffle=True)

    # Linear head for probing
    linear_head_sub = nn.Linear(2048, num_classes).to(device)
    optimizer_sub = optim.Adam(linear_head_sub.parameters(), lr=1e-3)
    criterion_sub = nn.CrossEntropyLoss()

    epochs_sub = 50
    train_accs_sub, val_accs_sub = [], []

    for epoch in range(epochs_sub):
        # ===================== TRAIN =====================
        linear_head_sub.train()
        total, correct = 0, 0
        for imgs, labels in subset_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            
            # Forward through encoder (no gradient)
            with torch.no_grad():
                feats = encoder(imgs)  # [batch_size, 2048]
            
            outputs = linear_head_sub(feats)  # [batch_size, num_classes]
            loss = criterion_sub(outputs, labels)
            
            optimizer_sub.zero_grad()
            loss.backward()
            optimizer_sub.step()
            
            _, preds = torch.max(outputs,1)
            correct += (preds==labels).sum().item()
            total += labels.size(0)
        train_accs_sub.append(correct/total)

        # ===================== VALIDATION =====================
        linear_head_sub.eval()
        total, correct = 0,0
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                feats = encoder(imgs)
                outputs = linear_head_sub(feats)
                _, preds = torch.max(outputs,1)
                correct += (preds==labels).sum().item()
                total += labels.size(0)
        val_accs_sub.append(correct/total)

    subset_train_accs.append(train_accs_sub[-1])
    subset_val_accs.append(val_accs_sub[-1])
    print(f"Linear Probe ({int(ratio*100)}% labels) | Train Acc: {train_accs_sub[-1]:.4f} | Val Acc: {val_accs_sub[-1]:.4f}")
```
